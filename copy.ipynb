# Imputer

_In data science, it's common for there to be missing values in a dataset. Let's see how we can create a class that will allow us to replace this missing value by the average of the values in the list_

1. _Create a class that we will call Imputer_.
2. _To simplify the exercise, we will only deal with lists for the moment._
3. _Our class will take an attribute that we will call list_.
4. _Create an avg() function that will first remove the missing value and then replace it with the average of the list._

```python
import numpy as np
```

```python
class Imputer:
    def __init__(self, listes):
        self.listes = listes

    def avg(self):
        # 1 - Nettoyer la liste en supprimant les None
        clean_list = [x for x in self.listes if x is not None]
        self.moyenne = float(np.mean(clean_list))
        # 2 - Remplacer les valeurs None par la moyenne
        return [self.moyenne if x is None else x for x in self.listes]
    
    def display(self):
        return f"Here's my initial list: {self.listes} and here's the list after imputation: {self.avg()}"

initial_list = [None, 2, 4, 6, None]
listes1 = Imputer(initial_list)
personpppprerAZ
display(listes1.display())
         
```

```python
class Imputer:
    def __init__(self, listes):
        self.listes = listes

    def avg(self):
        # 1 - Nettoyer la liste en supprimant les None
        clean_list = [x for x in self.listes if x is not None]
        self.moyenne = float(np.mean(clean_list))
        # 2 - Remplacer les valeurs None par la moyenne
        return [self.moyenne if x is None else x for x in self.listes]
    
    def med(self):
        # 1 - Nettoyer la liste en supprimant les None
        clean_list = [x for x in self.listes if x is not None]
        self.median = float(np.median(clean_list))
        # 2 - Remplacer les valeurs None par la moyenne
        return [self.median if x is None else x for x in self.listes]
    
    def display_avg(self):
        return f"Here is a list with the empty value replaced by the average value : {self.avg()}"
    def display_med(self):
        return f"Here is a list with the empty value replaced by the median value : {self.med()}"

initial_list = [None, 2, 3, 12, 5, 6, None]
listes1 = Imputer(initial_list)


print(f"here 's a list :{initial_list}")
print(listes1.display_avg())
print(f"here 's a list :{initial_list}")
print(listes1.display_med())      
```
